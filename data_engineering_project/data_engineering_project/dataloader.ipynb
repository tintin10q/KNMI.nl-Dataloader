```python
from tqdm import tqdm
def iter_nc_files():
    for file in glob.iglob("nc_files/*.nc"):
        db = netCDF4.Dataset(file)
        try:
            yield db
        finally:
            db.close()


counts = defaultdict(int)
file_count = 0
for db in iter_nc_files():
    for key in db.variables:
        var = db.variables[key]
        if var is None:
            counts[key] += 1
    file_count += 1
print(f"Iterated over {file_count} files. ")
print(f"There are {len(counts)} different keys with all the files")
print(f"Each key appearance counts:", counts)

```

```python
station_count = defaultdict(int)
station_indexes = defaultdict(set)

file_index = 0 

for db in iter_nc_files():
    file_index += 1
    stations = db.variables["station"]
    # print(stations[:])
    for station_index, station in enumerate(stations):
        station_count[station] += 1
        station_indexes[station].add((station_index, db.variables["wsik"]))
        
print("Read ", file_index, "files")
print(station_count)
print(station_indexes)

```

```python
def print_atters():
    counts = defaultdict(int)
    for file in tqdm(glob.glob("*.nc")):
        db = netCDF4.Dataset(file)
        for key in db.variables.keys():
            counts[key] += 1
        db.close()
    return counts
```

```python
def print_atter(attrbitue):
    for file in glob.glob("*.nc"):
        db = netCDF4.Dataset(file)
        print("\n\n",str(file),":")
        try: 
            for key in db.variables.keys():
                print(f"\t{key}: {getattr(db.variables[key],attrbitue)}", end=" ")
        except:
            print("Error")
            print(db.variables)
            return
        finally:
            db.close()
```

```python
from tomark import Tomark

def get_nc_filename() -> str:
    return next(glob.iglob("*.nc"))

def make_table():
    file = get_nc_filename()
    db = netCDF4.Dataset(file)
    result = []
    for key, value in db.variables.items():
        data =  db.variables[key]
        try:
            long_name = data.long_name
        except:
            long_name = None
        try:
            units = data.units
        except:
            units  = None
        try:
            print(key, data.comment)
        except:
            print(key,"no comment")
        # quantization
        row = {"name": data.name, "Long name": long_name, "size": data.size, "dimensions": data.dimensions, "units": units, "type": data.datatype,"scale": data.scale}
        result.append(row)
    db.close()
    return result



docs = make_table()
with open("table.table.md", "w+") as table_output:
    table_str = Tomark.table(docs)
    # print(table_str)
    table_output.write(table_str)
import os
```
